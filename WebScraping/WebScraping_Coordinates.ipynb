```python
# Import dependencies
import requests
import time
from config import api_key
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
from pprint import pprint
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
url = "https://www.latlong.net/latest-places.html"
browser.visit(url)
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
df = tables[0]
df.columns = ["Place Name","Latitude","Longitude"]
df.head()
```

```python
lat = df["Latitude"]
lat
```

```python
lon = df["Longitude"]
lon
```

```python
saving_to_list = []
for x in lat:
    for y in lon:
        lat_lng = f'https://api.airvisual.com/v2/nearest_city?lat={round(x,2)}&lon={round(y,2)}&key='+api_key
        time.sleep(12)# reach 'https://www.iqair.com/air-pollution-data-api/plans'
        lat_lng_response = requests.get(lat_lng).json()
        saving_to_list.append(lat_lng_response)
        pprint(lat_lng_response)
```

```python
len(saving_to_list)
```

```python
import json

with open('data.json', 'w') as f:
    json.dump(saving_to_list, f)
```
